```python
import os
from ultralytics import YOLO
from PIL import Image
import cv2
import matplotlib.pyplot as plt

# YOLO 모델 로드 (사전 학습된 모델 사용)
model = YOLO("yolov8n.pt")  # YOLOv8 Nano 모델 (빠르고 가벼움)

# 이미지 파일 리스트 수집
image_files = []
image_path = "/kaggle/input/korea-celebrities"

for dirname, _, filenames in os.walk(image_path):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        image_files.append(file_path)
```

```python
# 객체 검출 및 시각화 함수
def detect_objects(image_path):
    results = model(image_path)  # Object Detection 수행
    
    # 결과 시각화
    for result in results:        
        img = result.plot()  # 감지된 객체가 그려진 이미지
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

    # 검출된 객체 정보 출력    
    print(results[0].boxes.data)  # 바운딩 박스 좌표, 신뢰도, 클래스 ID 출력
```

```python
if image_files:
    for image_file in image_files:
        detect_objects(image_file)
```
